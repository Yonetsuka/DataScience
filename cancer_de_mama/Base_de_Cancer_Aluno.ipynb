```python
# 1)
df = pd.read_csv('breast-cancer-dataset.csv')
df = df.drop(columns=['S/N'])
df.to_csv('Excel_modificado.csv', index=False)
diferente = pd.read_csv('Excel_modificado.csv')
diferente
```

```python
# 2)
df = pd.read_csv('Excel_modificado.csv')
df = pd.get_dummies(df,columns=['Breast'], prefix = 'Breast')
df = pd.get_dummies(df,columns=['Metastasis'], prefix = 'Metastasis')
df = pd.get_dummies(df,columns=['Breast Quadrant'], prefix = 'Quadrant')
df = pd.get_dummies(df,columns=['History'], prefix = 'Histórico')
df = pd.get_dummies(df,columns=['Inv-Nodes'], prefix = 'Quadrant')
df = pd.get_dummies(df,columns=['Tumor Size (cm)'], prefix = 'Tamanho do tumor')
df = pd.get_dummies(df,columns=['Year'], prefix = 'Ano')
df = pd.get_dummies(df,columns=['Menopause'], prefix = 'Menopausa')
df.to_csv('Excel_modificado(1).csv', index=False)
outro = pd.read_csv('Excel_modificado(1).csv')
outro
```

```python
# 3)
df = pd.read_csv('Excel_modificado(1).csv')
y = df['Diagnosis Result']
X = df.drop(columns=['Diagnosis Result'])
```

```python
#4)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print("Acurácia do Modelo:", accuracy)
```
