```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
direcoes = [0,45,90,135,180]

def aplicar_rotacao(ponto,angulo):
    theta = np.radians(angulo)
    
    matriz_rotacao = np.array([[np.cos(theta), -np.sin(theta)],
                               [np.sin(theta), np.cos(theta)]])
    return np.dot(matriz_rotacao, ponto)

plt.figure(figsize = (10,5))
plt.title("Animação: Personagem olhando em diferentes direções ")

for angulo in direcoes:
    posicao_inicial = np.array([2*np.cos(np.radians(angulo)), 2*np.sin(np.radians(angulo))])
    posicao_rotacionada = aplicar_rotacao(posicao_inicial, angulo)
    plt.scatter(posicao_rotacionada[0], posicao_rotacionada[1], label = f'Direção: {angulo}', marker = 'o', s= 200)

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
